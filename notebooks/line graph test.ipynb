```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
music_df=pd.read_csv('../data/mlc_data.csv')
```

```python
music_df = music_df.rename(columns={'If ': 'ISRC'})
```

```python
music_df["ISRC"].fillna(" ", inplace = True)
music_df["Country Code"].fillna(" ", inplace = True)
music_df["Registrant Code"].fillna(" ", inplace = True)
music_df["Year of Reference"].fillna(" ", inplace = True)
music_df["Usage Period"].fillna(" ", inplace = True)
music_df["Streaming Platform (DSP)"].fillna(" ", inplace = True)
music_df["Streams"].fillna(" ", inplace = True)
music_df["Recording Artist"].fillna(" ", inplace = True)
music_df["Recording Label"].fillna(" ", inplace = True)
music_df["Recording Title"].fillna(" ", inplace = True)
music_df["Release Title"].fillna(" ", inplace = True)
music_df["Recording Duration (Seconds)"].fillna(" ", inplace = True)

```

```python
music_df=music_df.set_axis(["ISRC","Country","Registrant","Year","UsePeriod","Platform","Streams","Artist","Label","RecordingName","ReleaseName","Length","ListsWriters"], axis=1)
```

```python
music_df['Streams'] = music_df['Streams'].astype(str)
music_df['Streams']=music_df['Streams'].str.replace(',','')
music_df['Streams'] = music_df['Streams'].astype(int)
```

```python
music_df["Streams"] = music_df["Streams"]/1000
```

```python
music_df['UsePeriod'] = music_df['UsePeriod'].astype(str)
```

```python
music_df.info()
```

```python
month_streams = music_df.groupby(['UsePeriod', 'Platform'], as_index=False,).agg({'Streams':'sum'})

print(month_streams.head())
```

```python
month_streams.tail()
```

```python
Month = month_streams[month_streams['UsePeriod'].isin(['1/1/2021', '2/1/2021', '3/1/2021', '4/1/2021', '5/1/2021'])]
Amazon = Month[Month['Platform'].isin(['Amazon'])]
Spotify = Month[Month['Platform'].isin(['Spotify'])]
Apple = Month[Month['Platform'].isin(['Apple'])]
Pandora = Month[Month['Platform'].isin(['Pandora'])]
```

```python
Month.head()
```

```python
plt.subplots(figsize=(10,8))
plt.plot(Spotify.UsePeriod, Spotify.Streams, color="Blue")
plt.plot(Apple.UsePeriod, Apple.Streams, color="Green")
plt.plot(Pandora.UsePeriod, Pandora.Streams, color="Red")
plt.plot(Amazon.UsePeriod, Amazon.Streams, color="Orange")
plt.title('Streams by Month', fontsize=20, fontstyle='oblique')
plt.xlabel('Month', fontsize=20, fontstyle='oblique')
plt.ylabel('Streams (in 1000s)', fontsize=20, fontstyle='oblique')
plt.legend(["Spotify", "Apple", "Pandora", "Amazon"])
plt.show()
```
